```python
dati_sinistri = np.array([pt for pt in data if pt[1] < x[250]])
dati_destri = np.array([pt for pt in data if pt[1] >= x[250]])
plt.scatter(dati_sinistri[:, 1], dati_sinistri[:, 0])
plt.scatter(dati_destri[:, 1], dati_destri[:, 0])
```

```python
pt1 = (dati_sinistri[0, 1], dati_sinistri[0, 0])
pt2 = (dati_sinistri[1, 1], dati_sinistri[1, 0])
m = (pt1[1] - pt2[1]) / (pt1[0] - pt2[0])
x0 = pt1[0]
y0 = pt1[1]
plt.plot(np.arange(-30, -16, 0.001), m * (np.arange(-30, -16, 0.001) - x0) + y0)
plt.plot(dati_sinistri[:, 1], dati_sinistri[:, 0])
```

```python
pt1 = (dati_destri[0, 1], dati_destri[0, 0])
pt2 = (dati_destri[1, 1], dati_destri[1, 0])
m = (pt1[1] - pt2[1]) / (pt1[0] - pt2[0])
x0 = pt1[0]
y0 = pt1[1]
plt.plot(np.arange(-18, -2, 0.001), m * (np.arange(-18, -2, 0.001) - x0) + y0)
plt.plot(dati_destri[:, 1], dati_destri[:, 0])
```
